Convolution: $n^{[l]}=n^{[l-1]}+2p-f+1 $, if $p=(f-1)/2$ we get same convolution

**Strides** track the starting points and shift the filter with spacing

Convolution: $(n^{[l]}=n^{[l-1]}+2p-f)/2+1 $

## Volume Convolutions

take 3d tensors as input, use 3d filter (2d x #channels)

$$z^{[l]}=a^{[l-1]} * w^{[l]}+b^{[l]} $$
$$a^{[l]}=g(z^{[l]}) $$

## Complexity 

* $\displaystyle f^{[l]} $ filter size
* $\displaystyle p^{[l]} $ padding 
* $\displaystyle s^{[l]} $ stride
* $\displaystyle n_c^{[l]}$ number of filters

Computation: $O(N^2n^4)$ for each sample. N for input dimension, n for number of filters

## Pooling Layer 
* Reduce the size of the feature map
* Speeds up the optimization

Types of pooling: Max-pooling, Average-pooling, L2 norm. Normally, pooling is after convolution layer.

## Benefit of Convolutions (Pros)
* Parameter sharing. -
Filter can be useful in different parts of the input (image)

* Sparsity of connections -In each layer each output value depends only on samll number of inputs (local) -Translation invariance

## Challenges of Convolution (Cons?)

* Computational Complexity expensive - $O(N^2n^4)$

* Deeper Structure Needed - In each layer each output value depends only on small number of inputs (local)

## Classic Networks

* LeNet 5- Handwritten digits recognition
* AlexNet(7)- Object Classification
* VGG (16)- Recognition / Classification

**AlexNet Specifics**
- Much bigger than LeNet (60M parameters)
- ReLU
- Multiple GPUs
- Local Response Normalization (LRN)

**VGG (16) -> (19)**
- Conv: f=3, s=1, same
- Pool: f=2, s=2
- Order: CCP, CCP, CCCP, CCCP, CCCP, FFS
- Number filter: $2^6, 2^7, 2^8, 2^9, 2^9$
- Organized Structure, more layers, more parameters
